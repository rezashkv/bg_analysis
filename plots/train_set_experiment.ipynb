```python
import seaborn as sns
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt

test_datasets =      ['Original', 'Mixed-Same', 'Mixed-Rand']
vit_mixed_rand_acc = [21.85,      21.73,        17.60       ]
vit_original_acc =   [23.95,      23.28,        15.31       ]
res_mixed_rand_acc = [67.36,      62.59,        62.86       ]
res_original_acc =   [71.58,      60.72,        41.26       ]

plt.figure(figsize=(20, 11))
group_size = len(test_datasets)
barWidth = 1
spacing = group_size + barWidth*1.2
group1_x_locs = [i * barWidth for i in range(group_size)]
group2_x_locs = [i * barWidth + spacing for i in range(group_size)]
group3_x_locs = [i * barWidth + spacing * 2 for i in range(group_size)]
group4_x_locs = [i * barWidth + spacing * 3 for i in range(group_size)]

# colors = sns.color_palette(n_colors=group_size)
colors = ['navy', 'maroon', 'darkgoldenrod']
# plt.rcParams['font.family'] = 'CMU Serif'
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 28
sns.set_style("darkgrid", {"axes.facecolor": ".9"})

for i, dataset in enumerate(test_datasets):
    plt.bar(group1_x_locs[i], res_original_acc[i], color=colors[i], width=barWidth, label=dataset)
for i, dataset in enumerate(test_datasets):
    plt.bar(group2_x_locs[i], res_mixed_rand_acc[i], color=colors[i], width=barWidth)
for i, dataset in enumerate(test_datasets):
    plt.bar(group3_x_locs[i], vit_original_acc[i], color=colors[i], width=barWidth)
for i, dataset in enumerate(test_datasets):
    plt.bar(group4_x_locs[i], vit_mixed_rand_acc[i], color=colors[i], width=barWidth)

# Horizontal line at 11% accuracy
plt.axhline(y=11, color='black', linestyle='--', label='Random guess')

# Double-headed arrow indicating BG-Gap
bar = 2
arrow_x = group3_x_locs[bar] - barWidth/3  # The x location of the bar
arrow_y_start = vit_original_acc[bar]  # The top of the bar
arrow_y_end = vit_original_acc[bar-1]  # The top of the preceding bar
plt.annotate('', xy=(arrow_x, arrow_y_start), xytext=(arrow_x, arrow_y_end),
             arrowprops=dict(arrowstyle='<->', linewidth=3, color='black'))
plt.text(arrow_x + barWidth/6, (arrow_y_start+arrow_y_end)/2.2, 'BG-Gap', ha='left')

# We need to reorder the legend so the dashed line appears last
handles, labels = plt.gca().get_legend_handles_labels()
handles = handles[1:] + handles[:1]
labels = labels[1:] + labels[:1]
plt.legend(handles, labels, title='Test Dataset')

# Format axes, labels, and title
center = (barWidth * group_size / 2) - (barWidth / 2)
models = ['ResNet', 'ResNet (M-R trained)', 'ViT', 'ViT (M-R trained)']
tics = [center, center + spacing, center + spacing * 2, center + spacing * 3]
plt.xticks(tics, models)
plt.tick_params(axis='x', length=0, pad=10)
plt.xlabel('Model', fontweight='bold')
plt.ylabel('Accuracy', fontweight='bold')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y/100)))  # Format y axis as percentages
plt.title('Impact of Data Augmentation on BG-Gap - IN9', fontweight='bold', y=1.05)

plt.show()
```
