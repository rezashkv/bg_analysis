```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'CMU Serif'
plt.rcParams['font.size'] = 10

test_datasets = ['Mixed-Next', 'Mixed-Rand', 'Mixed-Same', 'Only-FG', 'Original']
mixed_rand_accuracies = [1, 21.85, 1, 17.6, 1, 1]
original_accuracies = [1, 1, 1, 1, 1, 1]

plt.figure(figsize=(10, 6))
group_size = len(test_datasets)
barWidth = 0.15
spacing = 1
group1_x_locs = [i * barWidth for i in range(group_size)]
group2_x_locs = [i * barWidth + spacing for i in range(group_size)]
colors = sns.color_palette(n_colors=group_size)

for i, dataset in enumerate(test_datasets):
    plt.bar(group1_x_locs[i], mixed_rand_accuracies[i], color=colors[i], width=barWidth, label=dataset)

for i, dataset in enumerate(test_datasets):
    plt.bar(group2_x_locs[i], original_accuracies[i], color=colors[i], width=barWidth)

plt.xlabel('Training Dataset', fontweight='bold')
plt.ylabel('Test Accuracy', fontweight='bold')
plt.legend(title='Test Dataset')
center = (barWidth * group_size / 2) - (barWidth / 2)
plt.xticks([center, center + spacing], ['Mixed-Rand', 'Original'])
plt.tick_params(axis='x', length=0, pad=10)

plt.show()
```
